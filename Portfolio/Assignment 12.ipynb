```python
import pandas as pd
import seaborn as sns

df = pd.read_csv('../datasets/adult.csv')
df.head()
```

```python
df.groupby('education').mean(numeric_only='true')
```

```python
sns.barplot(y="capital-gain", x="education", data=df)
```

### Conclusion
There is a clear difference between categories between capital gain per education group. This difference was to be expected because in general jobs that require higher eduction pay more.

```python
sns.barplot(y="capital-gain", x="race", data=df)
```

### Conclusion
They only concicive difference i can observe in the above barplot is that the capital gain of black people is lower than that of white people.

```python
sns.barplot(y="hours-per-week", x="gender", data=df)
```

### Conclusion
There is a significant difference between the amout of hours worked per week on average. Male work more than female according to this dataset.
