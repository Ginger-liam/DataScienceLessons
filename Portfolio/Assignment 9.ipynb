### Bivariate analasis

```python
import seaborn as sns

penguins = sns.load_dataset("penguins")
penguins.head()
```

All correlations using .corr()

```python
exampleDataCorrelations = penguins.corr(numeric_only=True)
exampleDataCorrelations.style.background_gradient(cmap='coolwarm', axis=None).format(precision=2)
```

### Summary of findings

Negative correlation

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate correlations for numerical columns
exampleDataCorrelations = penguins.corr(numeric_only=True)

# Visualize correlations with a heatmap
plt.figure(figsize=(15, 8))
sns.heatmap(exampleDataCorrelations, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap for Penguins Dataset')
plt.show()
```

### Conclusion
Some columns are correlated, like body mass and flipper size, this makes sense because as a penguin gets heavier they will need to displace more water to move themselves under water, and on land they can use the extra width to support their weight.

Bill depth and bill length are apparently not correlated, this information is usefull when training a model.
