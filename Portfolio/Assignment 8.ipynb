```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../../Datasets/Football/player_valuations.csv')

# Display the first few rows of the dataset to confirm it loaded correctly
print(df.head())
```

```python
# Step 1: Identify numerical columns
print(df.info())  # Check data types of all columns
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
print("Numerical columns:", numerical_columns)

# Step 2: Plot histograms for at least two numerical columns
columns_to_plot = numerical_columns[1:3]  # Select the first two numerical columns
for column in columns_to_plot:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[column], kde=True )
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```
