### Univariate analysis: Categorical data

```python
import pandas as pd

adults = pd.read_csv('../datasets/adult.csv')
adults.head()
```

```python
# Analyze the 'education' column
adults['education'].value_counts().plot(kind='bar')
```

Assumption: Most people have at least a high school education.
Finding: The bar plot shows 'HS-grad' is the most common education level.

```python
adults['income'].value_counts().plot(kind='bar')
```

Assumption: Most people earn less than 50k.     
Finding: The bar plot confirms this assumption.

```python
# Analyze the 'age' column
adults['hours-per-week'].plot(kind='hist', bins=20)
```

Assumption: Most adults work around 40 hours a week.
Finding: The histogram confirms most adults work about 40 hours per week.

```python
# Analyze the 'age' column
adults['age'].plot(kind='hist', bins=20)
```

Assumption: Most adults are between 25 and 50 years old.
Finding: The histogram confirms most ages fall in this range.
