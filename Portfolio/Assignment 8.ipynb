```python
import pandas as pd
import scipy.stats as st
```

```python
df = pd.read_csv('../Datasets/adult.csv')
ages = df['age'].dropna()
# Display the first few rows of the dataset to confirm it loaded correctly
df.head()
```

```python
mean_age = ages.mean()
sem_age = st.sem(ages)

confidence = .90
print(st.t.interval(confidence, len(ages)-1, loc=mean_age, scale=sem_age))

confidence = .95
print(st.t.interval(confidence, len(ages)-1, loc=mean_age, scale=sem_age))

confidence = .99
print(st.t.interval(confidence, len(ages)-1, loc=mean_age, scale=sem_age))

confidence = .9999
print(st.t.interval(confidence, len(ages)-1, loc=mean_age, scale=sem_age))
```

### Conclusion
As the confidence level increases, the confidence interval becomes wider. This occurs because a higher confidence level requires a broader range to ensure that the true mean is captured within the interval. In other words, to be more certain that the interval contains the true mean, we must allow for a larger possible range of values.
