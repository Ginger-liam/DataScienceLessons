### Univariate analysis
Univariate analysis of species from the standard seaborn dataset

```python
# import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load the penguins dataset
penguins = sns.load_dataset('penguins')
```

```python
# write the dataset data types
penguins.dtypes
```

```python
# print count per species
print(penguins['species'].value_counts())
```

```python
# display count of species in a bar plot
penguins['species'].value_counts().plot(kind='bar')
plt.title('Distribution of Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

### Summary
Most penguins in the dataset are of species Adelie. The majority are found on the Biscoe island.
